```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rate0 = 10000/np.array([26*60 + 30, 29*60 + 32, 36*60 + 38, 43*60 + 29, 45*60 + 8, 49*60 + 29])

rate15 = 10000/np.array([25* + 48, 28*60 + 16, 39*60 + 22, 52*60 + 47, 26*60 + 54, 31*60 + 19])

thr = np.array([3, 5, 10, 15, 17, 20]) #mv

fig, ax = plt.subplots(figsize = [7, 4])

ax.errorbar(thr, rate0, yerr = rate0*0.01, marker = "o", markersize = 2, label = "Channel 0, Gate 15ns")
ax.errorbar(thr, rate15, yerr = rate15*0.01, marker = "o", markersize = 2, label = "Channel 15, Gate 40ns")

ax.set_xlabel("Thr [mV]")
ax.set_ylabel("Dark Rate [Hz]")

ax.set_title("Misure Dark Rate 16/04/24, 10000 events each")

ax.legend()
ax.grid()

fig.savefig("./Dark_Rate_160424.pdf")
```
